```python
""" IMPORTANDO OS DADOS"""
import dask.dataframe as dd

df = dd.read_csv('/home/jovyan/data/central_west.csv')
df.head()
```

```python
""" FILTRANDO PARA SO DADOS DO DF """
import pandas as pd

distrito_federal = df[df['state'] == 'DF']
```

```python
int_cols = ['index', 'RADIACAO GLOBAL (Kj/m²)', 
            'UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)',
            'UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)',
            'UMIDADE RELATIVA DO AR, HORARIA (%)',
            'VENTO, DIREÇÃO HORARIA (gr) (° (gr))']

for col in int_cols:
    distrito_federal[col] = distrito_federal[col].astype('int32')
```

```python
category_cols = ['region', 'state', 'station', 'station_code']

for col in category_cols:
    distrito_federal[col] = distrito_federal[col].astype('category')
```

```python
distrito_federal.drop(columns=['index'])
```

```python
distrito_federal = distrito_federal.drop_duplicates()
distrito_federal = distrito_federal.dropna()
```

```python
distrito_federal = distrito_federal.drop(columns=['index'])
```

```python
distrito_federal.head()
```

```python
distrito_federal = distrito_federal.reset_index(drop=True)
```

```python
# Ajustando as colunas Data e Hora para o formato correto
distrito_federal['Data'] = distrito_federal['Data'].map_partitions(pd.to_datetime, format='%Y-%m-%d', errors='coerce')
distrito_federal['Hora'] = distrito_federal['Hora'].map_partitions(
    lambda x: pd.to_datetime(x, format='%H:%M:%S', errors='coerce').dt.time
)
```

```python
import boto3
import pandas as pd
from io import BytesIO

minio_client = boto3.client(
    's3',
    endpoint_url='http://localhost:9000',  
    aws_access_key_id='minioadmin',         
    aws_secret_access_key='minioadmin',   
)

bucket_name = 'climate-analysis'
object_key = 'distrito_federal_tratado.csv' 

df_pandas = distrito_federal.compute()

csv_buffer = BytesIO()
df_pandas.to_csv(csv_buffer, index=False)
csv_buffer.seek(0) 

minio_client.upload_fileobj(csv_buffer, bucket_name, object_key)
print(f"Arquivo enviado para {bucket_name}/{object_key}")
```
